### create a program that will calculate the Return on Investment(ROI) for a rental property.

```python
# The four categories needed to find ROI
# Income
# Expenses  #Utilities be its own dictionary
# Cash Flow
# Cash on Cash ROI
# each category should be a dictionary that prints out to the user


# Income function to calculate user's income
def income():
    income_dict = {}      #create empty dictionary, add in keys with values as inputs
    income_dict['Rental Income'] = int(input("What is your rental income?: "))
    income_dict['Laundry'] = int(input("What is your laundry income?: "))
    income_dict['Storage'] = int(input("What is your storage income?: "))
    income_dict['misc'] = int(input("What other incomes do you have?: "))
    income_dict['Total Income'] = income_dict['Rental Income'] + income_dict['Laundry'] + income_dict['Storage'] + income_dict['misc']
    global inc          # make varaible global so i can use it in another function
    inc = income_dict['Total Income']
    for i in income_dict:
        print(f'{i} : ${income_dict[i]}')


# Expense function to calculate user's expenses
def expenses():
    utilities_dict = {}           # seperate utilities dictionary to get total utilties value
    utilities_dict['Electric'] = int(input("How much is your electric bill per month?: "))
    utilities_dict['Water'] = int(input("How much is your water bill per month?: "))
    utilities_dict['Sewer'] = int(input("How much is your sewage bill per month?: "))
    utilities_dict['Garbage'] = int(input("How much is your garbage bill per month?: "))
    utilities_dict['Gas'] = int(input("How much is your gas bill per month?: "))
    utilities_dict['Other Utilities'] = int(input("How much is your other utilities per month?: "))
    utilities_dict['Total Utilities'] = utilities_dict['Electric'] + utilities_dict['Water'] + utilities_dict['Sewer'] + utilities_dict['Garbage'] + utilities_dict['Gas'] + utilities_dict['Other Utilities']
    for u in utilities_dict:
        print(f'{u} : ${utilities_dict[u]}')
        
    expenses_dict = {}     #expenses dictionary, add in keys with values as inputs, get utilites value from its dict
    expenses_dict['Tax'] = int(input("How much is your tax per month?: "))
    expenses_dict['Insurance'] = int(input("How much is your insurance per month?: "))
    expenses_dict['Utilities'] = utilities_dict.get('Total Utilities')     #plug in the value from the total utilites
    expenses_dict['HOA'] = int(input("How much is your Home Owners Association per month?: "))
    expenses_dict['Yardwork'] = int(input("How much are you paying to get your yard done per month?: "))
    expenses_dict['Vacancy'] = int(input("How much are you gonna place to the side for vacancy per month?: "))
    expenses_dict['Repairs'] = int(input("How much will you place to the side for repairs per month?: "))
    expenses_dict['Cap_Ex'] = int(input("How much is your Capital Expenditures per month?: "))
    expenses_dict['Property Management'] = int(input("How much is your property management per month?: "))
    expenses_dict['Mortgage'] = int(input("How much is your mortgage per month?: "))
    expenses_dict['Other Expenses'] = int(input("Are there any other expenses per month?: "))
    expenses_dict['Total Expenses per month'] = expenses_dict['Tax']+expenses_dict['Insurance']+expenses_dict['Utilities']+expenses_dict['HOA']+expenses_dict['Yardwork']+expenses_dict['Vacancy']+expenses_dict['Repairs']+expenses_dict['Cap_Ex']+expenses_dict['Property Management']+expenses_dict['Mortgage']+expenses_dict['Other Expenses']
    global ex            # make varaible global so i can use it in another function
    ex = expenses_dict['Total Expenses per month']
    for e in expenses_dict:
        print(f'{e} : ${expenses_dict[e]}')
    

# Cash flow function to calculate user's cash flow
def cash_flow():
    global monthcashflow                          # make varaible global so i can use it in another function
    total_month_cash_flow = inc - ex                   # using the global variables from income and expenses
    monthcashflow = total_month_cash_flow
    print(f'This is total monthly cash flow: ${total_month_cash_flow}') 
    

# ROI function to calculate user's Return on Investment
def ret_investment():
    roi_dict = {}      #empty dictionary for ROI, input is value
    roi_dict['Down Payment'] = int(input("How much was the down payment?: "))
    roi_dict['Closing Costs'] = int(input("How much was the closing costs?: "))
    roi_dict['Rehab Budget'] = int(input("How much was the rehabilitation on the property?: "))
    roi_dict['Misc'] = int(input("How much were other things invested in the property?: "))
    roi_dict['Total Investment'] = roi_dict['Down Payment'] + roi_dict['Closing Costs'] + roi_dict['Rehab Budget'] + roi_dict['Misc'] 
    for r in roi_dict:
        print(f'{r} : ${roi_dict[r]}')
    # the ROI 
    total_invest = roi_dict['Total Investment']
    annual_cash_flow = (monthcashflow * 12)    # need to make this a yearly figure so that it can be divided to investment  #grabing the global variable from cash_flow
    print(f'This is annual cash flow: {annual_cash_flow}')
    final_roi = (annual_cash_flow/total_invest) * 100   # make this figure into a percentage
    print(f'This is your ROI: {final_roi}%')


def run_ROI():
    income()
    expenses()
    cash_flow()
    ret_investment()


run_ROI()
```
